```python
#Original Bar graph: https://projects.fivethirtyeight.com/2020-endorsements/democratic-primary/

import matplotlib.pyplot as plt 
import numpy as np


objects = ('Joe Biden', 'Bernie Sanders', 'Elizabeth Warren', 'Kamala Harris', 'Mike Bloomberg', 'Cory Booker')
y_pos = np.arange(len(objects))
performance = [694,73,103,96,89,62]

plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('points')
plt.title('2020 Endorsement Party',fontsize = 18, fontweight = "bold", x = .51, y = 1.05)
plt.text(s="Which Democratic candidates are receiving the most support from prominent members of their party?", x=-100, y=5.75 )


plt.show()
```
